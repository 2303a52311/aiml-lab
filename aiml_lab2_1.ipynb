```python
import matplotlib.pyplot as plt
from queue import Queue # Import the Queue class

blood_grp =['A','B','O','AB']
A=[20,60,70,40]
plt.bar(blood_grp,A,color="lavender")
plt.title("Blood Group")
plt.xlabel("blood samples")
plt.ylabel("Frequency")
plt.show()

# Define the graph - This is an example, replace with your actual graph data
graph = {
    1: [2, 3],
    2: [4, 5],
    3: [5],
    4: [],
    5: []
}

print("The adjacenct List representing the graph is: ")
print(graph)

def bfs(graph,source):
  Q = Queue()
  visited_vertices = set()
  Q.put(source)
  visited_vertices.update({source})
  while not Q.empty():
    vertex = Q.get() # Changed '+' to '=' to correctly assign the value
    print(vertex,end="-->")
    for u in graph[vertex]:
      if u not in visited_vertices:
        Q.put(u)
        visited_vertices.update({u})

print("BFS traversal of graph with source 1 is: ")
bfs(graph,1)
```
